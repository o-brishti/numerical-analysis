## RK4 implementation to solve 1st order ODE

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t, N):
    return (- c * N)
```

```python
c = 0.132
N0 = 100

t0 = 0
tf = 12
h = 0.5
n = int((tf - t0)/h)

Nplot = np.zeros(n + 1)
tplot = np.zeros(n + 1)

Nplot[0] = N0
tplot[0] = t0

for i in range(n):
    N = Nplot[i]
    t = tplot[i]

    k1 = h * f(t, N)
    k2 = h * f(t + h/2, N + k1/2)
    k3 = h * f(t + h/2, N + k2/2)
    k4 = h * f(t + h, N + k3)

    tplot[i+1] = t + h
    Nplot[i+1] = N + (k1 + 2*k2 + 2*k3 + k4)/6

print(f"Amount of Substance left after 12 years = {Nplot[-1]} grams")
```

```python
plt.plot(tplot, Nplot, label = "N(t)")
plt.xlabel("Time")
plt.ylabel("Amount of Substance")
plt.title("Substance Decay using RK4")
plt.grid(True)
plt.legend()
plt.show()
```

## RK4 implementation to solve 2nd order ODE

```python
c = 0.5     # damping coefficient
k = 4.0     # spring constant
x0 = 1.0    # initial position
v0 = 0.0    # initial velocity
t0 = 0
tf = 10
h = 0.01
n = int((tf - t0)/h)
```

```python
def f(t, y):
    x, v = y
    dxdt = v
    dvdt = -c*v -k*x
    return np.array([dxdt, dvdt])
```

```python
tplot = np.zeros(n+1)
xplot = np.zeros(n+1)
vplot = np.zeros(n+1)

tplot[0] = t0
xplot[0] = x0
vplot[0] = v0

for i in range(n):
    t = tplot[i]
    y = np.array([xplot[i], vplot[i]])

    k1 = h * f(t, y)
    k2 = h * f(t + h/2, y + k1/2)
    k3 = h * f(t + h/2, y + k2/2)
    k4 = h * f(t + h, y + k3)

    tplot[i+1] = t + h
    ynew = y + (k1 + 2*k2 + 2*k3 + k4)/6
    xplot[i+1], vplot[i+1] = ynew
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(1, 3, 1)
plt.plot(tplot, xplot, label="Position vs Time")
plt.grid(True)
plt.xlabel("Time")
plt.ylabel("Position")
plt.title("Position vs Time of Damped Oscillator")

plt.subplot(1, 3, 2)
plt.plot(tplot, vplot, label="Velocity vs Time")
plt.grid(True)
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.title("Velocity vs Time of Damped Oscillator")

plt.subplot(1, 3, 3)
plt.plot(xplot, vplot, label="Velocity vs Position")
plt.grid(True)
plt.xlabel("Position")
plt.ylabel("Velocity")
plt.title("Phase Portrait")

plt.tight_layout()
plt.show()
```
